```python
# Create variable to calculate covariance
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
# Calculate covariance of Ratings and Installs
ratings, installs = df.Rating, df.Installs_Numeric
Cov(ratings, installs)
```

```python
# Calculate covariance of Ratings and Reviews
ratings, reviews = df.Rating, df.Reviews
Cov(ratings, reviews)
```

```python
# Create function to calculate Pearson Correlation
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
# Calculate Pearson Correlation of Ratings and Installs
Corr(ratings, installs)
```

```python
# Calculate Pearson Correlation of Ratings and Reviews
Corr(ratings, reviews)
```

```python
# Calculate the Correlation Coefficient of Ratings and Installs
np.corrcoef(ratings, installs)
```

```python
# Calculate the Correlation Coefficient of Ratings and Reviews
np.corrcoef(ratings, reviews)
```

```python
# Create function for the Spearman Correlation
def SpearmanCorr(xs, ys):
    xranks = pandas.Series(xs).rank()
    yranks = pandas.Series(ys).rank()
    return Corr(xranks, yranks)
```

```python
# Calculate Spearman Correlation of Ratings and Installs
SpearmanCorr(ratings, installs)
```

```python
# Calculate Spearman Correlation of Ratings and Reviews
SpearmanCorr(ratings, reviews)
```

```python
# Create class for the Difference in Means for the Null Hypothesis Test
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
# Calculate difference in means for Ratings of Paid vs. Free Apps for Null Hypothesis Test
data = free_apps.Rating.values, paid_apps.Rating.values
ht = DiffMeansPermute(data)
pvalue = ht.PValue()
pvalue
```

```python
# Plot the Null Hypothesis Test
ht.PlotCdf()
thinkplot.Show(xlabel = 'test statistic', ylabel='CDF')
```
